```python
class Mapper:
    def __init__(self):
        self.kotakc = kotakc  # Kotak instrument data (possibly DataFrame)
        self.kotakfno = kotakfno  # Kotak additional data (e.g., F&O numbers)
        self.zerodha_df = zerodha_df  # Zerodha DataFrame
        self.df21 = df21  # ICICI DataFrame 1
        self.df22 = df22  # ICICI DataFrame 2
        self.df23 = df23  # ICICI DataFrame 3 (and so on)
        self.df24 = df24
        self.df25 = df25       
        self.df1 = df1  # Fyers DataFrame 1
        self.df2 = df2  # Fyers DataFrame 2 (and so on)
        self.df3 = df3
        self.df4 = df4
        self.df5 = df5
        self.data = data  # Angel One data (possibly JSON)
        self.kotak = ["string","ShortName", self.kotakc, self.kotakfno]
        self.zerodha = ["string","name", self.zerodha_df]
        self.ICICI = ["string","shortname", self.df21, self.df22, self.df23, self.df24, self.df25]
        self.Fyers = ["number",13, self.df1, self.df2, self.df3, self.df4, self.df5]
        self.angel_one = ["json", self.data]
        self.others = {'kotak': self.kotak, "zerodha": self.zerodha, "ICICI": self.ICICI,"angel_one":["in",self.angel_one]}
    def find_common(self,input_script_name:str,output_script_name:str):
        if input_script_name=="angel_one":
            pass
        input_map=self.others[input_script_name]
        output_map=self.others[input_script_name]
        input_name=input_map[1]
        output_name=output_map[1]
        input_series=[]
        output_series=[]
        for i in range(2,len(input_map)):
            input_series.append(input_map[i][input_name])
        for i in range(2,len(input_map)):
            output_series.append(output_map[i][output_name])
        input_series=pd.concat(input_series,axis=0)
        output_series=pd.concat(output_series,axis=0)
        return input_series[input_series.isin(output_series) & output_series.isin(input_series)]

        
    def find_uncommon(self,input_script_name:str,output_script_name:str):
        if input_script_name=="angel_one":
            pass
        input_map=self.others[input_script_name]
        output_map=self.others[input_script_name]
        input_name=input_map[1]
        output_name=output_map[1]
        input_series=[]
        output_series=[]
        for i in range(2,len(input_map)):
            input_series.append(input_map[i][input_name])
        for i in range(2,len(input_map)):
            output_series.append(output_map[i][output_name])
        input_series=pd.concat(input_series,axis=0)
        output_series=pd.concat(output_series,axis=0)
        return input_series[~input_series.isin(output_series) & output_series.isin(input_series)]
        
    def output_script(self,input_script_name):
        input_map=self.others[input_script_name]
        return input_map[2:]
        
    
```
